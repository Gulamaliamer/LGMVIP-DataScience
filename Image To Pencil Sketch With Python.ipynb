# Importing necessary libraries to perform the task

```python
import cv2
import matplotlib.pyplot as plt
```

# Reading the image

```python
img = cv2.imread('ab-de-villiers.jpg',0)
img
```

# Converting the image into gray

```python
plt.imshow(img, cmap='gray')
```

# Stating the length of image max & min

```python
img.max()
```

```python
img.min()
```

# Inverting the gray scale image 

```python
inv = 255-img
```

```python
plt.imshow(inv, cmap='gray')
```

# Image Blur by using Gaussian Blur Function

```python
Blur = cv2.GaussianBlur(inv, ksize=(21,21), sigmaX=0, sigmaY=0)
```

```python
plt.imshow(Blur, cmap='gray')
```

# Plotting the Pencil Sketch of Image

```python
dodge = lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
blended = dodge(img, Blur)
```

```python
plt.imshow(blended, cmap='gray')
```

# Save Sketch Zip File

```python
cv2.imwrite('imgsketch.jpg', blended)
```
